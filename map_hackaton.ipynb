```python
import folium

# Create the map
m = folium.Map(location=[51.505, -0.09], zoom_start=13)

# Add a marker
folium.Marker([51.505, -0.09], popup="Marker 1").add_to(m)

# Add a polygon
folium.Polygon(
    locations=[
        [51.505, -0.09],
        [51.51, -0.1],
        [51.5, -0.08]
    ],
    color="blue",
    fill=True,
    fill_color="cyan"
).add_to(m)

# Display the map in Jupyter Notebook
m

```

```python
import json

# Extract GeoJSON data from the map
geojson_data = m.to_json()  # Serialize Folium map to JSON string

# Save GeoJSON to a file
with open("map_data.json", "w") as f:
    json.dump(json.loads(geojson_data), f)
```
